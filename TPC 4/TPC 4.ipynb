```python
import random

def criaLista(N):
    n = 1
    global lista
    lista = []
    while n <= N:
        lista.append(random.randrange(1,101))
        n = n + 1
    return(lista)

def leLista(N):
    global lista
    lista = []
    while N > 0:
        k = int(input("Escolha um número"))
        lista.append(k)
        N = N - 1
    return(lista)

def somaLista(N):
    global lista
    resultado = 0
    for n in N:
        resultado = resultado + n
    return resultado

def mediaLista(N):
    global lista
    resultado = 0
    a = 0
    for n in N:
        resultado = resultado + n
        a = a + 1
        media = resultado / a 
    return media

def maiorLista(N):
    global lista
    maior = N[0]
    for n in N:
        if n > maior:
            maior = n
    return maior

def menorLista(N):
    global lista
    menor = N[0]
    for n in N:
        if n < menor:
            menor = n
    return menor

def estaOrdenadaC(N):
    global lista
    if len(N) <= 1:
        return True
    for n in range(1, len(N)):
        if N[n] <= N[n - 1]:
            return False
    return True

def estaOrdenadaD(N):
    global lista
    if len(N) <= 1:
        return True
    for n in range(1, len(N)):
        if N[n] >= N[n - 1]:
            return False
    return True

def procura(N, elem):
    global lista
    indice = -1
    if elem in N:
        indice = N.index(elem) + 1
    return indice

escolha = -1
while escolha != 0:
    escolha = int(input("""Introduza o número correspondente à opção que quer escolher:
    (1) Criar Lista 
    (2) Ler Lista
    (3) Soma
    (4) Média
    (5) Maior
    (6) Menor
    (7) estaOrdenada por ordem crescente
    (8) estaOrdenada por ordem decrescente
    (9) Procura um elemento
    (0) Sair"""))
    if escolha == 1:
        print(criaLista(int(input("Quantos números quer na lista?"))))
    elif escolha == 2:
        print(leLista(int(input("Quantos números quer na lista?"))))
    elif escolha == 3:
        print(somaLista(lista))
    elif escolha == 4:
        print(mediaLista(lista))
    elif escolha == 5:
        print(maiorLista(lista))
    elif escolha == 6:
        print(menorLista(lista))
    elif escolha == 7:
        print(estaOrdenadaC(lista))
    elif escolha == 8:
        print(estaOrdenadaD(lista))
    elif escolha == 9:
        print(procura(lista, int(input("Escolha um número"))))    

```
